### In this notebook we will use the raw data to create a simulated different dataset to test datadrift.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("..\\data\\01_raw\\Hotel_H2.csv")
df.tail()
```

```python
data = df[-10000:]
```

```python
data.index = [i+10000 for i in data.index]
```

We can not mess a lot with dates because it requires a lot of fiddling to get accurate dates according to week number, weekends, etc. So we will stay in 2017. We will change little things in the dataset so we can just test data drift and see how it affects scores. We will see an influx of British people and their ADR will increase.

```python
data.columns
```

```python
data
```

```python
data.loc[data.index[5000:],'Country'] = 'GBR'
data.loc[data.index[5000:],'ADR'] += 50
```

```python
len(data.index[5000:])
```

```python
data.IsCanceled.value_counts()
```

Let's also pretend they cancel more (75% of the time).

```python
data.loc[data.index[5000:],'IsCanceled'] = [np.random.choice([0, 1], p=[0.25, 0.75]) for i in range(5000)]

```

```python
data.IsCanceled.value_counts()
```
